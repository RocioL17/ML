```python
import matplotlib.pyplot as plt
import numpy as np

# Simulación de la selección aleatoria de características
features = ['X1', 'X2', 'X3', 'X4', 'X5', 'X6']
selected_features = [np.random.choice(features, size=2, replace=False) for _ in range(5)]

print("Características seleccionadas en cada nodo:")
for i, subset in enumerate(selected_features):
    print(f"Nodo {i+1}: {subset}")

# Gráfico para ilustrar la randomización
plt.figure(figsize=(8, 4))
plt.bar(range(len(features)), [1] * len(features), color='gray', label='Todas las características')
for i, subset in enumerate(selected_features):
    indices = [features.index(f) for f in subset]
    plt.scatter(indices, [1] * len(indices), color='red', label=f'Nodo {i+1}' if i == 0 else None)
plt.xticks(range(len(features)), features)
plt.title("Randomización de Características en Random Forest")
plt.legend()
plt.show()
```
