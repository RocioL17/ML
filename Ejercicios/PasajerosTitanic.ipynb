Este dataset contien datos sobre los pasajeros del Titanic y no, no están ni Rose ni Jack. La idea es la de obtener un informe estadístico descriptivo sencillo. La salida será un archivo HTML que podrás descargar de la carpeta Files.:

```python
# Descargar el dataset
!wget https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv -O titanic.csv
import pandas as pd
from ydata_profiling import ProfileReport

# Cargar el dataset
df_titanic = pd.read_csv('titanic.csv')

# Visualizar los primeros registros
print("Primeros registros del dataset:")
print(df_titanic.head())

# Contar la cantidad de registros
print("\nCantidad de registros:", len(df_titanic))

# Análisis descriptivo básico
print("\nDescripción estadística del dataset:")
print(df_titanic.describe(include='all'))

# Generar un informe de análisis con YData Profiling
profile_titanic = ProfileReport(df_titanic, title="Análisis Descriptivo - Titanic")
profile_titanic.to_file("titanic_report.html")
print("\nInforme de análisis generado como 'titanic_report.html'")
```
