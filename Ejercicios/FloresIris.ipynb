```python
# Instalar librerías necesarias
!pip install pandas ydata-profiling

# Importar librerías
import pandas as pd
from ydata_profiling import ProfileReport

# Descargar el dataset
!wget https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv -O iris.csv

# Cargar el dataset
df_iris = pd.read_csv('iris.csv')

# Visualizar los primeros registros
print("Primeros registros del dataset:")
print(df_iris.head())

# Contar la cantidad de registros
print("\nCantidad de registros:", len(df_iris))

# Análisis descriptivo básico
print("\nDescripción estadística del dataset:")
print(df_iris.describe(include='all'))

# Generar un informe de análisis con YData Profiling
profile_iris = ProfileReport(df_iris, title="Análisis Descriptivo - Iris")
profile_iris.to_file("iris_report.html")
print("\nInforme de análisis generado como 'iris_report.html'")
```
